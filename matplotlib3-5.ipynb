```python
#03. Matplotlib 축 레이블 설정하기
#xlabel(), ylabel(), axis()

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
#여백 지정하기: labelpad, pt단위

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-Label', labelpad=15) 
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

```python
#폰트 설정
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel("X",fontdict={'family': 'serif', 'color': 'b',
                         'weight': 'bold', 'size': 14})
plt.ylabel('Y-Axis', labelpad=20, fontdict={'family': 'fantasy', 'color': 'deeppink',
                                            'weight': 'normal', 'size': 'xx-large'})
plt.show()
```

```python
#위치 지정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-axis',loc='right')
plt.ylabel('Y-axis',loc='center')
plt.show()
```

```python
#04. Matplotlib 범례 표시하기
#label 문자열을 지정하고 legend() 함수 호출

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Price ($)')
plt.legend()
plt.show()
```

```python
#위치 지정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='AAAAAAA')
#plt.legend(loc=(0,0))
plt.legend(loc = [0.5,0.5])
#plt.legend(loc=(1,1))
#plt.legend(loc='lower right')

plt.show()
```

```python
#열 개수 지정하기, 폰트사이즈, 범례 꾸미기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Price ($)')
plt.plot([1,2,3,4],[3,5,9,7], label = 'Demand (#)')
plt.legend(ncol = 1, fontsize = 14, frameon=True, shadow=True)
#plt.legend(loc='best',ncol = 2, fontsize = 14)
plt.show()
```

```python
#05. Matplotlib 축 범위 지정하기
#xlim(), ylim()

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlim([1,2])
plt.ylim(1,100)
plt.show()
```

```python
#axis()

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.axis([0,5,0,20])
plt.show()

```

```python
#옵션 지정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.axis('scaled')
#plt.axis('square')
plt.show()
```

```python
#축 범위 얻기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('square')
#axis_range = plt.axis()
print(axis_range)

plt.show()
```
