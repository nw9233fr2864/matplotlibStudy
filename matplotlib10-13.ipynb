```python
#10. Matplotlib 축 스케일 지정하기
#xscale(), yscale()

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = x**3

plt.plot(x,y)
plt.xscale('symlog')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,100)
y = np.exp(x)

plt.plot(x,y)
#plt.yscale('linear')
plt.yscale('log')

plt.show()
```

```python
#11. Matplotlib 여러 곡선 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.2)

plt.plot(x, x, 'r--', x, x**2, 'bo', x, x**3, 'g-.')
plt.show()
```

```python
#12 Matplotlib 그리드 설정하기
#grid()

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.2)

plt.plot(x,x)
plt.grid(True)

plt.show()
```

```python
#축 지정하기
#axis, color, alpha, linestyle, which

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.2)

plt.plot(x,x)
plt.grid(True, axis='y', linestyle='--', which = 'major')

plt.show()
```

```python
#13. Matplotlib 눈금 표시하기
#xticks(), yticks(), tick_params()
#python list, numpy array를 인자로 사용 가능
#label 지정도 가능

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.xticks([0, 1, 2], labels=['Jan','Feb','Mar'])
plt.yticks(np.arange(1, 6))

plt.show()

```

```python
#눈금 스타일 설정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.tick_params(axis='x', direction = 'inout', length = 3, pad = 6,
               labelsize = 14, labelcolor = 'green', top = True)

plt.show()
```
