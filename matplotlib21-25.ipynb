```python
#24. Matplotlib 여러 개의 그래프 그리기
#subplot()

import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1) #nrows, ncols, index
plt.plot(x1,y1)
plt.title('1')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('2')
plt.xlabel('time')
plt.ylabel('Undamped')

#plt.tight_layout()
plt.show()
```

```python
#축 공유하기

import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1)
ax2 = plt.subplot(2,1,2, sharex=ax1)######
plt.plot(x2,y2)

plt.tight_layout()
plt.show()
```

```python
#25. Matplotli 컬러맵 설정하기
#컬러바 나타내기

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
arr = np.random.standard_normal((8,100))

plt.subplot(2,2,1)
#plt.scatter(arr[0],arr[1],c=arr[1], cmap = 'spring')
plt.scatter(arr[0],arr[1],c=arr[1])
plt.spring()
plt.title('spring')
plt.colorbar()

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.summer()
plt.title('summer')
plt.colorbar()

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.autumn()
plt.title('autumn')
plt.colorbar()

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.winter()
plt.title('winter')
plt.colorbar()

plt.tight_layout()
plt.show()
```
