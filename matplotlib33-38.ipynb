```python
#33. Matplotlib 이중 Y축 표시하기
#twinx()

import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['font.size'] = 12

x = np.arange(0,3)
y1 = x+1
y2 = -x-1

fig, ax1 = plt.subplots()
ax1.plot(x,y1)

ax2 = ax1.twinx()
ax2.plot(x,y2)

plt.show()
```

```python
#축 레이블 표시하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,3)
y1 = x+1
y2 = -x-1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.plot(x,y1)

ax2 = ax1.twinx()
ax2.set_ylabel('2st Y-Axis')
ax2.plot(x,y2)

plt.show()
```

```python
#범례 표시하기

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.plot(x, y1, color='green', label='1st Data')
ax1.legend(loc='upper right')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis')
ax2.plot(x, y2, color='deeppink', label='2nd Data')
ax2.legend(loc='lower right')


#line1 = ax1.plot(x, y1, color='green', label='1st Data')
#line2 = ax2.plot(x, y2, color='deeppink', label='2nd Data')
#lines = line1 + line2
#labels = [l.get_label() for i in lines]
#ax1.legend(lines, labels, loc='upper right')

plt.show()
```

```python
#34. Matplotlib 두 종류의 그래프 그리기
#set_zorder()

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

fig, ax1 = plt.subplots()

ax1.plot(x, y1, 'r')
ax1.set_ylim(0,18)
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.bar(x, y2)

plt.show()
```

```python
#그래프 순서 지정하기
#z-order가 작을수록 먼저 그려지고, 밑그림이 됨

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

fig, ax1 = plt.subplots()

ax1.plot(x,y1, 'r')
ax2 = ax1.twinx()
ax2.bar(x,y2)

ax1.set_zorder(ax2.get_zorder()+10)
ax1.patch.set_visible(False)

plt.show()
```
