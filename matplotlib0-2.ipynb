```python
#01. mathploylib 기본사용

import matplotlib.pyplot as plt

plt.plot([1,2,3,4]) #y-value
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16]) #x-y graph
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16],"ro")
#plt.plot([1,2,3,4],[1,4,9,16],"b-")
plt.axis([0,6,0,20]) #xmin, xmax, ymin, ymax
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0,5,0.2)

plt.plot(t,t,'r--',t,t*82,'bs', t,t**3,'g^') #s: square, ^: triangle
plt.show()
```

```python
#02. Matplotlib 숫자 입력하기
#plot()의 괄호안에 튜플이나 Numpy array형태로 입력할 수 있음
#y-value만 입력, 즉 하나의 배열만 입력하였을 때 x는 0부터 1씩 증가함
#dictionary 사용법

import matplotlib.pyplot as plt

data_dic = {'d_x': [1,2,3,4,5], 'd_y': [2,3,5,10,8]}

plt.plot('d_x','d_y', data = data_dic)
plt.show()
```
