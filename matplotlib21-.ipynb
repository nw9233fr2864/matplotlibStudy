```python
#24. Matplotlib 여러 개의 그래프 그리기
#subplot()

import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1) #nrows, ncols, index
plt.plot(x1,y1)
plt.title('1')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('2')
plt.xlabel('time')
plt.ylabel('Undamped')

#plt.tight_layout()
plt.show()
```

```python
#축 공유하기

import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1)
ax2 = plt.subplot(2,1,2, sharex=ax1)######
plt.plot(x2,y2)

plt.tight_layout()
plt.show()
```
