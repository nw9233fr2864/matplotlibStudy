```python
#rc Params 사용하기

import matplotlib.pyplot as plt

plt.style.use('bmh')
plt.rcParams['figure.figsize'] = (6,3) #그래프 이미지의 크기
plt.rcParams['font.size'] = 12 #폰트의 크기
plt.rcParams['lines.linewidth'] = 5
plt.rcParams['lines.linestyle'] = '--'

plt.plot([1,2,3,4],[2,3,5,10])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'inout'
plt.rcParams['ytick.direction'] = 'in'

plt.plot([1,2,3,4],[2,3,5,10])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['xtick.major.size'] = 7
plt.rcParams['ytick.major.size'] = 7
plt.rcParams['xtick.minor.visible'] = True
plt.rcParams['ytick.minor.visible'] = True

plt.plot([1,2,3,4],[2,3,5,10])

plt.show()

#etc.
```

```python
#29 Matplotlib 이미지 저장하기
#savefig()

import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0,5)
x2 = np.linspace(0,2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
#plt.show()
plt.savefig('savefig_default.png')
```

```python
#dpi 설정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])

plt.savefig('savefig_default.png', dpi=1000) #dots per inch
```

```python
#facecolor 설정하기
#edgecolor 설정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5)
y = x**x

plt.figure(linewidth=5) #기본 linewidth가 0이므로 설정해 주어야함

plt.plot(x,y)
plt.savefig('savefig_edgecolor.png', facecolor='#eeeeee', edgecolor='blue')
```
