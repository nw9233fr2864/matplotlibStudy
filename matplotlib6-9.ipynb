```python
#06. Matplotlib 선 종류 지정하기
#포맷 문자열을 사용하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], '-', color = 'C2', label = 'Solid') # - -- : -.
plt.show()
```

```python
#linestyle로 지정하기

import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
#plt.tight_layout()

plt.show()
```

```python
#linestyle을 튜플로 지정하기

import matplotlib.pyplot as plt

# (0, (크기, 간격, [크기, 간격]))
plt.plot([1, 2, 3], [4, 4, 4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))') 
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

```python
#선 끝 모양 지정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], linewidth=10, solid_capstyle = 'round')
plt.plot([1,2,3,4],[3,5,9,7], linewidth=10, solid_capstyle = 'butt')

plt.show()
```

```python
#07 Matplotlib 마커 지정하기

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10],'ks--') #이전의 선 종류와 동시에 지정

plt.show()
```

```python
#09. Matplotlib 그래프 영역 채우기
#fill_between(), fill_betweenx(), fill()

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3],alpha=0.1)
plt.fill_betweenx(y[1:3],x[1:3])
plt.show()
```

```python
#두 그래프 사이 영역 채우기

import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [2,3,5,6]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill_between(x[1:4],y2[1:4],y1[1:4])

plt.show()
```

```python
#다각형 영역 채우기

import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill([1.9,1.9,3.1,3.1],[1,4,6,3])

plt.show()
```
