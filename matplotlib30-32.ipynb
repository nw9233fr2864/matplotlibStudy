```python
#31. Matplotlib 객체 지향 인터페이스 2

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,5)

fig, ax = plt.subplots(2,2,sharex=True, sharey=True, squeeze=True)
#squeeze는 2D array Axes로 변환, True면 1D로 변환 가능, False면 무조건 2D로 변환, default: True
ax[0][0].plot(x,np.sqrt(x))
ax[0][1].plot(x,x)
ax[1][0].plot(x,-x+5)
ax[1][1].plot(x, np.sqrt(-x+5))

plt.show()
```

```python
#스타일 설정하기, 제목과 범례 표시하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,5)

fig, ax = plt.subplots(2,2,sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x,np.sqrt(x), label = 'y=np.sqrt(x)')
ax[0][0].set_title('Graph1')
ax[0][0].legend()
ax[0][1].plot(x,x, label = 'y=x')
ax[0][1].set_title('2')
ax[1][0].plot(x,-x+5, linewidth = 5)
ax[1][1].plot(x, np.sqrt(-x+5))

plt.show()
```

```python
#32. Matplotlib 축 위치 조절하기
#spines()
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (6,3)
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots()

ax.set_title("Mean Squared Error", pad=20)
ax.axis([-3,3,0,3])
ax.set_xticks([-3,-2,-1,0,1,2,3])
ax.set_yticks([1,2,3])

ax.spines['left'].set_position('center')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position(('data',0))

x = np.linspace(-3,3,100)
ax.set_xlabel('y_true - y_pred', labelpad = 10)
ax.plot(x,x**2)

plt.show()
```
