```python
#14. Matplotlib 타이틀 설정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.2)

plt.plot(x,x)
plt.title('11111111', loc = 'right', pad = 20)
title_font =  {
    'fontsize': 16,
    'fontweight': 'bold'
}

plt.title('22222222',fontdict = title_font, loc = 'left')
plt.show()
```

```python
#타이틀 얻기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
title_left = plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)

print(title_left.get_position())
print(title_left.get_text())

print(title_right.get_position())
print(title_right.get_text())

plt.show()
```

```python
#15 Matplotlib 수평선/수직선 표시하기
#axhline(), axvline(), hlines(), vlines()
#axh/vline은 0에서 1사이의 값, h/vlines는 value

import matplotlib.pyplot as plt

plt.xlim(-20,10)
plt.ylim(0,10)
plt.plot([1,3],[3,4])
plt.axhline(3,0.1,1,color = 'lightgray')
plt.hlines(2,1,2)
plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

```python
#16. Matplotlib 막대(Bar) 그래프 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years) #x축의 눈금 레이블

plt.show()
```

```python
#색상 지정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x,values, color = colors)
#plt.bar(x,values, color = 'y')
plt.show()
```

```python
#막대 폭 지정하기
#스타일 꾸미기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, width=0.4, edgecolor = 'lightgray', #default wide = 0.8
        linewidth = 5,tick_label = years)
plt.show()
```

```python
#17. Matplotlib 수평 막대 그래프 그리기
#막대 높이 지정
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y,values, height = 0.4)
plt.yticks(y,years)
plt.show()
```
